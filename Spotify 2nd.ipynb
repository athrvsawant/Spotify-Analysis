```python
import numpy as np 
import pandas as pd 
import matplotlib.pyplot as plt
import seaborn as sns 
%matplotlib inline 

df_genre = pd.read_csv(r'D:\sample data\Spotify\archive (1)\SpotifyFeatures.csv')
df_genre
```

# Top 7 Genres

```python
df_genre.head(7)
```

# Duartion of Songs w.r.t Different Genres

```python
plt.title("Duration of the Songs in Different Genres")
sns.color_palette("rocket",as_cmap=True)
sns.barplot(y='genre', x = 'duration_ms', data = df_genre)
plt.xlabel("Duration in MilliSeconds")
plt.ylabels("Genres")
```

# Top 5 Genres by Popularity

```python
sns.set_style(style = "darkgrid")
plt.figure(figsize=(10,5))
famous = df_genre.sort_values("popularity", ascending = False).head(8)
sns.barplot(y = 'genre', x = 'popularity', data = famous).set(title = 'Top 4 Genres by Popularity')
```
